```python
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python

#Open the shine page on automated chrome brower
driver.get(" https://www.amazon.in/")
```

```python
# Enter  PRODUCT NAME
product = driver.find_element(By.XPATH, "html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
product.send_keys('guitar')


```

```python
# CLICK ON SEARCH
Search = driver.find_element(By.XPATH,"html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
Search.click()
```

```python
# scraping  brand name from given page.

brand=[]
start =0
end = 3
for page in range(start,end):
     brand_tags = driver.find_elements(By.XPATH, '//span[@class="a-size-base a-color-base"]')
for i in brand_tags[0:10]: 
    brand.append(i.text)
    next_button = driver.find_elements(By.XPATH,'//a[@href="/s?k=guitar&page=2&crid=24G5VTSQY2IBK&qid=1707981317&sprefix=guitar%2Caps%2C379&ref=sr_pg_1"]') 
#scraping product name from given page. 
product_name =[]
product_tags =driver.find_elements(By.XPATH,'//span[@class="a-size-base-plus a-color-base a-text-normal"]')
for i in product_tags[0:10]:
    product_name.append(i.text)
# scraping expected delivery from given page.  
expected_delivery =[]
Delivery_tags = driver.find_elements(By.XPATH,'//div[@data-cy="delivery-recipe"]')
for i in Delivery_tags[0:10]:
    expected_delivery.append(i.text)
# scraping price from given page.     
price =[] 
price_tags = driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
for i in price_tags[0:10]:
    price.append(i.text)
#scraping avabilability in given page.    
avabilability =[]
avabilable_tags = driver.find_elements(By.XPATH,'//span[@class="a-size-base a-color-base puis-bold-weight-text"]')
for i in avabilable_tags[0:10]:
    avabilability.append(i.text)
#scraping page url in given page.      
page_url=[]
url = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
for i in url[0:10]:
    page_url.append(i.get_attribute('href'))
    
    
       
       
    
```

```python
len(brand),len(product_name),len(expected_delivery),len(price),len(avabilability),len(page_url)
```

```python
page_url
```

```python
import pandas as pd
df = pd.DataFrame({'BrandName':brand,'ProductName':product_name,'Price':price,'Avabilability':avabilability,'Expected_Delivery':expected_delivery,'Page_Url':page_url})
df
```
