```python
driver = webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python


rank=[]
rank_tags = driver.find_elements(By.XPATH,'//td[@align="center"]')
for i in rank_tags[0:10]:
    rank.append(i.text)


name=[]
name_tags = driver.find_elements(By.XPATH,'//a[@class="mw-redirect"]')
for i in name_tags[0:10]:
    name.append(i.text)


artist_name =[]
artist_tags =driver.find_elements(By.XPATH,'//a[@class="mw-redirect"]')
for i in artist_tags[0:10]:
    artist_name.append(i.get_attribute("href"))

    
upload =[]
date_tags = driver.find_elements(By.XPATH,'//td[@align="right"]')
for i in date_tags[0:10]:
    upload.append(i.text)
    
    
     
views=[]
view_tags = driver.find_elements(By.XPATH,'//td[@align="center"]')
for i in view_tags[0:10]:
    views.append(i.text)

```

```python
len(rank),len(name),len(artist_name),len(upload),len(views)
```

```python
import pandas as pd 
df = pd.DataFrame({'Rank':rank,'Name':name, 'ArtistName':artist_name,'UploadDate':upload,'Views':views})
df
```
