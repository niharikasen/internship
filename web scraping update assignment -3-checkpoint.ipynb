```python

#Open the shine page on automated chrome brower
driver.get(" https://www.amazon.in/")
```

```python
# Enter  PRODUCT NAME
product = driver.find_element(By.XPATH, "html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input")
product.send_keys('guitar')


```

```python
# CLICK ON SEARCH
Search = driver.find_element(By.XPATH,"html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input")
Search.click()
```

```python
# scrape produt url
product_url = []
start =0
end =3    
for i in range (start, end):
    url = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
for i in url[0:20]:
    product_url.append(i.get_attribute("href"))
nxt_button = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')  
nxt_button.click()
time.sleep(2)

```

```python
len(product_url)
```

```python
for url in product_url:
    driver.get(url)
    time.sleep(2)
    
brand =[] 

try :
        brand = driver.find_elements(By.XPATH, '//span[@class="a-size-base a-text-bold"]')
        Brand.append(brand.text)
        
except NoSuchElementException:
                Brand.append('_')     
        
        
product=[]
try:
    product =driver.find_elements(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
    Product.append(product.text)   
        
except NoSuchElementException:
                Product.append('_')
        
expected_delivery =[]
try:
    expected_delivery = driver.find_elements(By.XPATH,'//div[@data-cy="delivery-recipe"]')
    delivery.append(expected_delivery.text)
except NoSuchElementException:
                delivery.append('_')        
Price =[] 
try:
    Price = driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
    Price_tag.append(Price.text)
except NoSuchElementException:
    Price_tag.append('_')        
            
avabilability =[]
try:
    avabilability = driver.find_elements(By.XPATH,'//span[@class="a-size-base a-color-base puis-bold-weight-text"]')
    avaliable.append(avabilability.text)
except NoSuchElementException:
    avaliable.append('_')
           
    
```

```python
import pandas as pd
df = pd.DataFrame({'BrandName':brand,'Product':product,'Price':Price_tag,'Avabilability':avabilable,'Expected_Delivery':delivery})
df
```

```python
len(brand),len(smartphone_name),len(screen_size),len(price),len(primary_camere),len(storage_ram),len(storage_rom),len(page_url)
```

```python
import pandas as pd
df= pd.DataFrame({'Brand':brand,'Smartphone':smartphone_name,'ScreenSize':screen_size,'Price':price,'PrimaryCamere':primary_camere,'RAM':storage_ram,'Rom':storage_rom,'PageURL':page_url})
df
```
